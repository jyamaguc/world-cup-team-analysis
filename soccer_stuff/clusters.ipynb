```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\jyama\project_stuff\soccer_stuff\Fifa_world_cup_matches_cleaned.csv')
df.head()
```

```python
df_1 = df.filter(regex='1$')
df_2 = df.filter(regex='2$')
```

```python
df_1.columns = df_1.columns.str[:-1]
df_1.head()
```

```python
df_2.columns = df_2.columns.str[:-1]
df_2.head()
```

```python
df = df_stacked = pd.concat([df_1, df_2])
df.head()
```

```python
df.shape
```

```python
df['possession team'] = df['possession team'].str.rstrip('%').astype('float') / 100
df.head()
```

```python
teams_df = df['team']
teams_list = list(df['team'].unique())
teams_list
```

```python
rows_list=[]
for item in teams_list:
    # Finding the rows in the DataFrame that contain the same string as the current item in the list
    matches = df[df['team'].str.contains(item)]
    # Getting the mean of each column in the matches DataFrame
    mean = matches.mean()
    # Creating a single row DataFrame from the mean Series
    mean_df = pd.DataFrame(mean).T
    # Adding a new column to the single row DataFrame to keep the item name
    mean_df = mean_df.assign(team=item)
    # Adding the single row DataFrame to the list
    rows_list.append(mean_df)

# Concatenating the matching rows into a new DataFrame
new_df = pd.concat(rows_list)
new_df = new_df.reset_index(drop=True)
new_df.head()
```

```python
print(len(teams_list), new_df.shape)
```

```python
new_df = new_df.drop('team', axis = 1)
new_df.head()
```

```python
#PCA on new_df
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

# Fitting and transforming the DataFrame to get the transformed data
df_transformed = pca.fit_transform(new_df)
```

```python
import matplotlib.pyplot as plt
x = df_transformed[:,0]
y = df_transformed[:,1]
plt.scatter(x,y)
```

```python
from sklearn.cluster import KMeans

sse = []

# Loop over a range of number of clusters
for k in range(1, 10):
    # Fit the KMeans model to the data with k clusters
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(new_df)
    
    # Compute the sum of squared distances for each data point and its corresponding cluster centroid
    sse.append(kmeans.inertia_)

# Plot the sum of squared distances for each number of clusters
plt.plot(range(1, 10), sse)
plt.xlabel("Number of Clusters")
plt.ylabel("Sum of Squared Distances")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(new_df)

# Predict the cluster labels for each data point
labels = kmeans.predict(new_df)

# Add the cluster labels back to the original dataframe
new_df["cluster"] = labels

# Plot the data points and color them based on their cluster
new_df['team']= teams_list
new_df
```
