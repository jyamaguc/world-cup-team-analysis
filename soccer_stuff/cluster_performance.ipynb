```python
df = pd.read_csv(r'C:\Users\jyama\project_stuff\soccer_stuff\Fifa_world_cup_matches_cleaned.csv')
df.head()
```

```python
df_games = df[['team1','team2','number of goals team1','number of goals team2']]
df_games.head()
```

```python
df_games.rename(columns={
    "number of goals team1": "goals1",
    "number of goals team2": "goals2"
}, inplace=True)
df_games.head()
```

```python
result = []

# Loop through each row in the dataframe
for index, row in df_games.iterrows():
    if row["goals1"] > row["goals2"]:
        result.append(1)
    elif row["goals1"] < row["goals2"]:
        result.append(2)
    else:
        result.append(0)
```
